The arithmetic error occurs when an error is encountered during numeric calculations in Python. This includes  OverflowError, Zerodivision Error and Floating point error.

```python
# Answer 3
j = 5.0

try:
    for i in range(1, 1000):
        j = j**i
        print(j)
except OverflowError as e:
    print("trying to handel Overflow error",e)
```
